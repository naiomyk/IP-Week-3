Data Cleaning

```
#cleaning telcom_dataset for duplicates
Telcom_dataset
#Rename the mispelling in product to match the other telcom datasets
Telcom_dataset.rename(columns = {'PRODUTC':'PRODUCT','DATETIME':'DATE_TIME','DW_A_NUMBER_INT':'DW_A_NUMBER','DW_b_NUMBER':'DW_B_NUMBER'}, inplace=True)
Telcom_dataset
#checking for duplicates
Telcom_dataset.duplicated()
Telcom_dataset_pro=Telcom_dataset.drop_duplicates()
Telcom_dataset_pro
```

```
#cleaning telcom_dataset2 for duplicates
Telcom_dataset2
#Telcom_dataset3.rename(columns = {'D}, inplace=True)
#checking for duplicates and dropping
Telcom_dataset2.duplicated()
Telcom_dataset2_pro=Telcom_dataset2.drop_duplicates()
Telcom_dataset2_pro
```

```
#cleaning telcom_dataset3 for duplicates
Telcom_dataset3
Telcom_dataset3.rename(columns = {'SIET_ID':'SITE_ID','DW_A_NUMBER_INT':'DW_A_NUMBER','DW_b_NUMBER':'DW_B_NUMBER', 'CELLID':'CELL_ID'}, inplace=True)
#checking for duplicates and dropping them
Telcom_dataset3.duplicated()
Telcom_dataset3_pro=Telcom_dataset3.drop_duplicates()
Telcom_dataset3_pro
```

```
#Drop unncessary columns

Telcom_dataset_clean = Telcom_dataset_pro.drop(columns = ['COUNTRY_A' ,	'COUNTRY_B' ,'CELL_ON_SITE'])
Telcom_dataset_clean
```

```
Telcom_dataset2_clean = Telcom_dataset2_pro.drop(columns = ['COUNTRY_A' ,	'COUNTRY_B' ,'CELL_ON_SITE'])
Telcom_dataset2_clean
```

```
Telcom_dataset3_clean = Telcom_dataset3_pro.drop(columns = ['COUNTRY_A' ,	'COUNTRY_B' ,'CELL_ON_SITE','DW_A_NUMBER','DW_B_NUMBER'])
Telcom_dataset3_clean
```

Cells_geo Data cleaning

```
#rename necessary columns fo ease in understanding
cells_geo.rename(columns = {'Unnamed: 0':'INDEX' ,'VILLES':'CITY','DECOUPZONE':'GEOGRAPHICAL'}, inplace=True)
cells_geo

#identify and remove duplicates in the cells_geo datasets

cells_geo.duplicated()
cells_geo.drop_duplicates()
```

```
#remove the null values
cells_geo.isnull()
cells_geo_no_missing = cells_geo.dropna()
cells_geo_no_missing
```

```
#clean to remove unnecessary columns

cells_geo_clean = cells_geo_no_missing.drop(columns = ['INDEX', 'LONGITUDE', 'LATITUDE', 'ZONENAME', 'STATUS', 'AREA', 'GEOZONE'])
cells_geo_clean
```

DATA PREPARATION


```
#appending the telcom datasets to have one 

Telcom_combined_clean=pd.concat([Telcom_dataset_clean,Telcom_dataset2_clean,Telcom_dataset3_clean], ignore_index = 1)

Telcom_combined_clean.tail()
```

```
#Merging the two tables to get the final dataframes

Final_dataset = Telcom_combined_clean.merge(cells_geo_clean, on=['CELL_ID'], how='outer')
Final_dataset
```
